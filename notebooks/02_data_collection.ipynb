```python
import fastf1
import pandas as pd
import os

# Enable cache to avoid re-downloading data
fastf1.Cache.enable_cache(r"C:\Users\Siddharth\OneDrive\Desktop\F1 Project\data\raw\cache")
print("✅ Cache enabled!")
```

```python
#Single Race test
session=fastf1.get_session(2024,'Belgium','R')
session.load(telemetry=False, weather=True, messages=False)

results=session.results
print(results.columns.tolist())
print(results.head())
```

```python
# Weather data test
weather= session.weather_data
print(weather.columns.tolist())
print(weather.head())
print("\nRainfall detected:", weather['Rainfall'].any())
print("Avg Track Temp:   ", weather['TrackTemp'].mean().round(1))
print("Avg Air Temp:     ", weather['AirTemp'].mean().round(1))
```

```python
# Qualifying test
quali = fastf1.get_session(2024, 'Belgium', 'Q')
quali.load(telemetry=False, weather=False, messages=False)

quali_results = quali.results[['DriverId', 'Q1', 'Q2', 'Q3']].copy()

# Convert lap times to seconds for easier math later
for col in ['Q1', 'Q2', 'Q3']:
    quali_results[col] = pd.to_timedelta(quali_results[col]).dt.total_seconds()

print(quali_results.head())
```
